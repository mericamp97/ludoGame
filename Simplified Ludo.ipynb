# This is a simplified version of Ludo. 
Each player has only 1 piece. The computer plays alone.

# Initialize the board
The board has 68 positions.

```python
#El taulell té 68 posicions. 
taulell = []
i = 1
while i < 69:
    taulell.append(i)
    i += 1

print(taulell)
```

# Initialize each player's piece
Each player begins in a different position in the board.

```python
#Cada color té una posició diferent on començar

pieces = ['blue', 'green', 'yellow', 'red']

init_blue = 5
init_green = 22
init_yellow = 39
init_red = 56

initial_positions = [init_blue, init_green, init_yellow, init_red]
init_dic = {}
i = 0
for piece in pieces:
    init_dic[piece] = initial_positions[i]
    i += 1

print(init_dic)
```

# Creating Routes
Every player begins at different points in the board, so they will have different routes to follow to reach their final destination. All players need to do 64 steps, but they all begin at different points in the board.

```python
#Cada peça té el seu recorregut

recorreguts = {'blue':{}, 'green':{}, 'yellow':{}, 'red':{}}

for color in pieces:
    position = init_dic[color]
    count = 1
    pos_dictionary = {}
    while count < 65:
        pos_dictionary[count] = position
        count += 1
        if position + 1 > 68:
            position = 1
        else:
            position += 1
    recorreguts[color] = pos_dictionary
    
print(recorreguts)
    
```

# Let's play!
Every player will play until someone reaches position 64. That player is the winner. If a player gets in the same board field as another piece, that piece is killed and the playing piece advances 20 positions!

```python
#Cada jugador juga fins que arriba a la seva posició 64

import random
playing_positions = {'blue':0, 'green':0, 'yellow':0, 'red':0}
playing_scores = {'blue':1, 'green':1, 'yellow':1, 'red':1}

while not (playing_scores['blue'] == 64 or playing_scores['green'] == 64 or playing_scores['yellow'] == 64 or playing_scores['red'] == 64): 
    for color in pieces:
        tir = random.randint(1,6)
        print('The color ', color, 'moves ',tir,' forward')
        reverse_positions = {value: key for key, value in playing_positions.items() if key != color}
        if playing_scores.get(color,0) + tir > 64:
            playing_scores[color] = 64
            playing_positions[color] = recorreguts[color][playing_scores[color]]
            print(color, ' arrived home!')
            print('\n')
            continue
        else:
            playing_scores[color] = playing_scores.get(color,0) + tir
            playing_positions[color] = recorreguts[color][playing_scores[color]]
        print(playing_scores)
        print(playing_positions)
        while playing_positions[color] in reverse_positions.keys():
            #The killed piece returns home
            print('\n')
            print('!!!!!')
            print('The color ',reverse_positions[playing_positions[color]], 'is dead!')
            print('!!!!!')
            print('\n')
            playing_scores[reverse_positions[playing_positions[color]]] = 1
            playing_positions[reverse_positions[playing_positions[color]]] = init_dic[reverse_positions[playing_positions[color]]]
            #The playing piece advances 20 positions
            if playing_scores.get(color,0) + 20 > 64:
                playing_scores[color] = 64
                playing_positions[color] = recorreguts[color][playing_scores[color]]
                print(color, ' arrived home!')
                print(playing_scores)
                print(playing_positions)
                print('\n')
                continue
            else:
                playing_scores[color] = playing_scores.get(color,0) + 20
                playing_positions[color] = recorreguts[color][playing_scores[color]]
                print(playing_scores)
                print(playing_positions)
                print('\n')
        
dic_sorted = sorted(playing_scores.items(), key = lambda x:x[1], reverse = True)

print('\n')
print('The winner is ', dic_sorted[0][0])
print('The second position is for ', dic_sorted[1][0])
print('The third position is for ', dic_sorted[2][0])
print('The complete loser is ', dic_sorted[3][0])
```
